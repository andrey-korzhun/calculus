__3.__ Численно найти хотя бы один действительный корень системы нелинейных уравнений:
### $$\begin{cases}
   x^2-y^2+3xy^3-2x^2y^2+2x-3y-5=0 \\
   3y^3-2x^2+2x^3y-5x^2y^2+5=0
 \end{cases}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

from sympy import plot_implicit, symbols, Eq
from scipy.interpolate import splev, splrep
from scipy.optimize import fsolve

warnings.filterwarnings("ignore")


x, y = symbols('x y')

p1 = plot_implicit(
    Eq(x**2 - y**2 + 3*x*y**3 - 2*(x**2)*(y**2) + 2*x - 3*y - 5, 0),
    (x, -4, 4), (y, -2, 2), line_color='blue', show=False)
p2 = plot_implicit(
    Eq(3*y**3 - 2*x**2 + 2*y*x**3 - 5*(x**2)*(y**2) + 5, 0),
    (x, -4, 4), (y, -2, 2), line_color='green', show=False)

p1.extend(p2)

p1.show()
```

```python
def equations(p):
    x, y = p
    e1 = x**2 - y**2 + 3*x*y**3 - 2*(x**2)*(y**2) + 2*x - 3*y - 5
    e2 = 3*y**3 - 2*x**2 + 2*y*x**3 - 5*(x**2)*(y**2) + 5
    return (e1, e2)

x, y = np.round(fsolve(equations, (-4, 0)), 8)
solution = np.round(equations((x, y)), 6)[0] == np.round(equations((x, y)), 6)[1]

print(*np.round((x, y), 8), solution)
```

__4*.__ Численно найти все $5$ действительных корней.

```python
sol = set()

for coord_x in range(-5, 5):
    for coord_y in range(-3, 3):
        coord = (coord_x, coord_y)
        x, y = np.round(fsolve(equations, coord), 8)
        if np.round(equations((x, y)), 6)[0] == np.round(equations((x, y)), 6)[1]:
            sol.add((x, y))

sol
```

__5*.__ Даны две функции $y_{1}=f(x)$ и $y_{2}=g(x)$. Известно, что:

### $$f'(x)=\frac{1}{2}g(x)$$

### $$g'(x)=2-2f(x)$$

### $$f(0)=0$$

### $$g(0)=1$$

Восстановить функции $y_{1}=f(x)$ и $y_{2}=g(x)$ с помощью формулы:

### $$f(x+\Delta x) \approx f(x)+f'(x) \cdot \Delta x$$

на участке $[0, 20]$

В качестве решения построить график этих функций в одной системе координат.


```python
N = 1000
L = 20

h = L / N
x = np.arange (0, L, h)
x = np.append(x, L)
f = [0]
g = [1]

while(N != 0):
    f = np.append(f, f[len(f) - 1] + 0.5 * g[len(f) - 1] * h)
    g = np.append(g, g[len(g) - 1] + (2 - 2 * f[len(g) - 1]) * h)
    N -= 1
x2 = np.linspace(0, L, 200)

plt.figure(figsize=(10, 5))
plt.axis('equal')
plt.grid()

spl_f = splrep(x, f)
f2 = splev(x2, spl_f)
plt.plot(x2, f2, label="y1 = f(x)")

spl_g = splrep(x, g)
g2 = splev(x2, spl_g)
plt.plot(x2, g2, label="y2 = g(x)")

plt.legend()
```

__6**.__ Найти все корни уравнения:
    
### $$f(x)=g(x)$$

на участке $[0, 20]$

Где $y_{1}=f(x)$ и $y_{2}=g(x)$ - функции из предыдущего решения.

```python
def equations(p):
    x3 = p
    return splev(x3, spl_f) - splev(x3, spl_g)

sol = set()

for x3 in np.arange (0, L, h):
    x3, info, ier, mesg =  fsolve(equations, x3, full_output=True)
    if (ier == 1 and x3 > 0 and x3 < L): sol.add(int(x3 * 10e8) / 10e8)

sol
```

```python
plt.figure(figsize=(10, 5))
plt.axis('equal')
plt.grid()
plt.plot(x2, f2, label="y1 = f(x)")
plt.plot(x2, g2, label="y2 = g(x)")
plt.legend()

sol_list = list(sol)
f3 = splev(sol_list, spl_f)

plt.scatter(sol_list, f3, c='red')

for i in range(len(f3)):
    print(f'{sol_list[i]:.8f}, {f3[i]:.8f}')
```
