__8**.__ С помощью метода наименьших квадратов (МНК) подобрать значения параметров $a$ и $b$ для сигмоидальной функции $$y=\frac{1}{1+e^{-(ax+b)}}$$ 

по заданному распределению. Построить график.

```python
np.random.seed(42)

xi = np.random.uniform(-8, 8, 200)
xi.sort()

yi = 1/(1+np.exp(-np.random.uniform(0.5, 1.5, 200)*xi-np.random.uniform(-1, 1, 200)))
plt.plot(xi,yi, marker="o", ls="")

x = np.linspace(-8, 8, num=200, endpoint=True)
y = 1 / (1+np.exp(-1*x))

plt.plot(x, y)
```

```python
U, f, x, y, a, e, b = symbols('U f x y a e b')
f = 1 / (1+e**(-a*x+b))
f_a = diff(f, a)
f_b = diff(f, b)

display(f_a)
display(f_b)
```

```python
def equations(p):
    A, B = p
    return ((-2*xi*np.exp(-A*xi-B) * (yi-1/(1+np.exp(-A*xi-B))) / (1+np.exp(-A*xi-B))**2).sum(), 
            (-2*np.exp(-A*xi-B) * (yi-1/(1+np.exp(-A*xi-B))) / (1+np.exp(-A*xi-B))**2).sum())

plt.plot(xi,yi, marker="o", ls="")

A, B =  fsolve(equations, (1, 0))
display(Markdown(f'### a = {A:.3f}, b = {B:.3f}'))

x = np.linspace(-8, 8, 200)
yn = 1/(1+np.exp(-A*x-B))

plt.plot(x, yn)

sko3 = np.sqrt(np.sum(np.square(yi - yn)) / 199)
print("Относительная погрешность аппроксимации:", sko3 / np.max(yn) * 100)

plt.show()
```

### Попробуем построить полином...

```python
xi.sort()

xp = np.array(xi).flatten()

ya3 = np.polyval(np.polyfit(xp, yi, 5), xp)
plt.plot(xi, ya3, c='Magenta')
plt.scatter(xi, yi)
sko3 = np.sqrt(np.sum(np.square(yi - ya3)) / 199)
print("Относительная погрешность аппроксимации:", sko3 / np.max(ya3) * 100)
```

### Построим зависимость относительной погрешности от степени полинома

```python
N = 8
sko_list = []
for i in range(1, N):
    ya_n = np.polyval(np.polyfit(xp, yi, i), xp)
    sko_n = np.sqrt(np.sum(np.square(yi - ya_n)) / 199)
    sko_list.append(sko_n / np.max(ya_n) * 100)
    
plt.plot(range(1, N), sko_list)
plt.xlabel("Степень регрессии")
plt.ylabel("Относительная погрешность")
```
