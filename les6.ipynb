```python
import sympy as sp

x = sp.Symbol('x', real=True)

f = (72 * x) - x**2

fprime = f.diff(x)

solution = [(xx, f.subs(x, xx)) for xx in sp.solve(fprime, x)]
solution[0][0]
```
